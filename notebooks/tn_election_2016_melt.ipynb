```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file StateGeneralbyPrecinctNov2016.xlsx
tn_2016=pd.read_excel('../data/2016_election/StateGeneralbyPrecinctNov2016.xlsx',
                      usecols=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE','RNAME1','PARTY1','PVTALLY1',
                             'RNAME2','PARTY2','PVTALLY2','RNAME3','PARTY3','PVTALLY3','RNAME4','PARTY4','PVTALLY4',
                             'RNAME5','PARTY5','PVTALLY5','RNAME6','PARTY6','PVTALLY6','RNAME7','PARTY7','PVTALLY7',
                             'RNAME8','PARTY8','PVTALLY8','RNAME9','PARTY9','PVTALLY9','RNAME10','PARTY10','PVTALLY10'])
tn_2016
```

```python
tn_2016.info()
```

```python
# merge entries in all RNAME columns into new column "name"
tn_2016_namemelt=tn_2016.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['RNAME1','RNAME2','RNAME3','RNAME4','RNAME5','RNAME6','RNAME7','RNAME8','RNAME9','RNAME10'],value_name="name")
# drop column "variable" created by .melt
tn_2016_namemelt=tn_2016_namemelt.drop(['variable'],axis=1)                          
tn_2016_namemelt.head(9405)
```

```python
# merge entries in all PARTY columns into new column "party"
tn_2016_partymelt=tn_2016.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['PARTY1','PARTY2','PARTY3','PARTY4','PARTY5','PARTY6','PARTY7','PARTY8','PARTY9','PARTY10'],value_name="party")
# drop column "variable" created by .melt
tn_2016_partymelt=tn_2016_partymelt.drop(['variable'],axis=1)
tn_2016_partymelt.head(9405)
```

```python
# merge entries in all PVTALLY columns into new column "vote_count"
tn_2016_pvtallymelt=tn_2016.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['PVTALLY1','PVTALLY2','PVTALLY3','PVTALLY4','PVTALLY5','PVTALLY6','PVTALLY7','PVTALLY8','PVTALLY9','PVTALLY10'],value_name="vote_count")
# drop column "variable" created by .melt
tn_2016_pvtallymelt=tn_2016_pvtallymelt.drop(['variable'],axis=1)
tn_2016_pvtallymelt.head(9405)
```

```python
# add "party" and "vote_count" columns to tn_2016_namemelt
party=tn_2016_partymelt['party']
pvtally=tn_2016_pvtallymelt['vote_count']
tn_2016_namemelt=tn_2016_namemelt.join(party).join(pvtally)
tn_2016_namemelt.head(3)
```

```python
tn_2016_namemelt.info()
```

```python
# lower case column names
tn_2016_namemelt.columns=tn_2016_namemelt.columns.str.lower()
tn_2016_namemelt.head(2)
```

```python
# change null values in "party" column to "no party"
tn_2016_namemelt['party'].fillna("No party", inplace = True)
tn_2016_namemelt.head(2)
```

```python
tn_2016_namemelt.info()
```

```python
# change null values in "vote_count" column to "0"
tn_2016_namemelt['vote_count'].fillna(0)
```

```python
tn_2016_namemelt.info()
```

### 16824 rows have a vote count of 0. Drop those rows

```python
# create data frame tn_2016_pres for presidental election
tn_2016_pres_int=tn_2016_final[tn_2016_final['officename'].str.match('United States President')]
tn_2016_pres_int=tn_2016_pres_int.reset_index(drop=True)
tn_2016_pres_int
```

```python
tn_2016_pres_int.head(20)
```

### Add column that has total vote count by count for use in calculating county vote percentages

```python
# group county precincts into county per candidate, in descending order of votes within county
tn_2016_pres_int=tn_2016_pres_int.groupby(['officename','county','name','party'], as_index=False)['vote_count'].sum()
tn_2016_pres=tn_2016_pres_int.sort_values(by=['county','vote_count'],ascending=False).reset_index(drop=True)
tn_2016_pres.head(20)
```

```python
# create new dataframe "county_vote" to use to find sum of votes per county
county_vote=tn_2016_pres.loc[0:,['county','vote_count']]
county_vote
```

```python
#sum county votes
county_vote_sum=county_vote.groupby(['county'])['vote_count'].sum()
county_vote_sum
```

```python
# use "county_vote_sum" to add column "sum_county_vote" to tn_2016_pres correlating with value in each row
tn_2016_pres['sum_county_vote']=tn_2016_pres['county'].map(county_vote_sum)
tn_2016_pres.head(20)
```

```python
# create "pct_county_vote" column for tn_2016_pres
tn_2016_pres['pct_county_vote']=round(tn_2016_pres['vote_count']/tn_2016_pres['sum_county_vote']*100,2)
tn_2016_pres
```

```python
# how many parties?
tn_2016_pres['party'].value_counts()
```

### Create dataframe "tn_2016_county_pres_pct"  with percent values per party per county

```python
# create new dataframe with reduced columns values only
tn_2016_pct_county=tn_2016_pres.loc[0:,['county','party','pct_county_vote']]
tn_2016_pct_county
```

```python
# create new dataframe with Republican values only
tn_2016_rep_pct=tn_2016_pct_county[tn_2016_pct_county['party']== "Republican"].reset_index(drop=True)
tn_2016_rep_pct
```

```python
# rename 'pct_county_vote' to 'rep_pct' and drop "party" column to prepare for merge 
tn_2016_rep_pct=tn_2016_rep_pct.rename(columns={'pct_county_vote':'rep_pct'},inplace=False).drop(['party'], axis=1)
tn_2016_rep_pct
```

```python
# create new dataframe with Democratic values only
tn_2016_dem_pct=tn_2016_pct_county[tn_2016_pct_county['party']== "Democratic"].reset_index(drop=True)
tn_2016_dem_pct
```

```python
# rename 'pct_county_vote' to 'dem_pct' and drop "party" to prepare for merge 
tn_2016_dem_pct=tn_2016_dem_pct.rename(columns={'pct_county_vote':'dem_pct'},inplace=False).drop(['party'], axis=1)
tn_2016_dem_pct
```

```python
# create new dataframe with non Republican or Democratic values only
tn_2016_other_pct=tn_2016_pct_county[(tn_2016_pct_county['party']=="Independent")|(tn_2016_pct_county['party']=="No party")].reset_index(drop=True)
tn_2016_other_pct
```

```python
# sum by county
tn_2016_other_pct=tn_2016_other_pct.groupby(['county'],as_index=False)['pct_county_vote'].sum()
tn_2016_other_pct
```

```python
# rename 'pct_county_vote' to 'other_pct' and drop "party" to prepare for merge 
tn_2016_other_pct=tn_2016_other_pct.rename(columns={'pct_county_vote':'other_pct'})
tn_2016_other_pct
```

```python
# join tn_2016_rep, tn_2016_dem_pct and tn_2016_other_pct to form tn_2016_county_pres_pct
dem=tn_2016_dem_pct['dem_pct']
other=tn_2016_other_pct['other_pct']
tn_2016_county_pres_pct=tn_2016_rep_pct.join(dem).join(other)
tn_2016_county_pres_pct
```

```python

tn_2016_pct_county['rep_pct']=tn_2016_pct_county['county'].map(tn_2016_rep_pct.set_index('county')['rep_pct'])
tn_2016_pct_county.head(25)
```

```python
# add 'rep_vic_margin' column to show Republican margin over other parties per county
tn_2016_pct_county['repub_vic_margin']=tn_2016_pct_county['rep_pct']-tn_2016_pct_county['pct_county_vote']
tn_2016_pct_county
```

```python
# Confirm negative numbers
tn_2016_pct_county['repub_vic_margin'].unique()
```

```python
# join tn_2016_margin 'rep_vic_margin' column to tn_2016_pres
rep_pct=tn_2016_pct_county['repub_vic_margin']
tn_2016_pres=tn_2016_pres.join(rep_pct)
tn_2016_pres.head(3)
```
