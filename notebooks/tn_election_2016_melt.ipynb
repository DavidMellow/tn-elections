```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file StateGeneralbyPrecinctNov2016.xlsx
tn_2016=pd.read_excel('../data/2016_election/StateGeneralbyPrecinctNov2016.xlsx',
                      usecols=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE','RNAME1','PARTY1','PVTALLY1',
                             'RNAME2','PARTY2','PVTALLY2','RNAME3','PARTY3','PVTALLY3','RNAME4','PARTY4','PVTALLY4',
                             'RNAME5','PARTY5','PVTALLY5','RNAME6','PARTY6','PVTALLY6','RNAME7','PARTY7','PVTALLY7',
                             'RNAME8','PARTY8','PVTALLY8','RNAME9','PARTY9','PVTALLY9','RNAME10','PARTY10','PVTALLY10'])
tn_2016
```

```python
tn_2016.info()
```

```python
# merge entries in all RNAME columns into new column "name"
tn_2016_namemelt=tn_2016.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['RNAME1','RNAME2','RNAME3','RNAME4','RNAME5','RNAME6','RNAME7','RNAME8','RNAME9','RNAME10'],value_name="name")
# drop column "variable" created by .melt
tn_2016_namemelt=tn_2016_namemelt.drop(['variable'],axis=1)                          
tn_2016_namemelt.head(9405)
```

```python
# merge entries in all PARTY columns into new column "party"
tn_2016_partymelt=tn_2016.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['PARTY1','PARTY2','PARTY3','PARTY4','PARTY5','PARTY6','PARTY7','PARTY8','PARTY9','PARTY10'],value_name="party")
# drop column "variable" created by .melt
tn_2016_partymelt=tn_2016_partymelt.drop(['variable'],axis=1)
tn_2016_partymelt.head(9405)
```

```python
# merge entries in all PVTALLY columns into new column "vote_count"
tn_2016_pvtallymelt=tn_2016.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['PVTALLY1','PVTALLY2','PVTALLY3','PVTALLY4','PVTALLY5','PVTALLY6','PVTALLY7','PVTALLY8','PVTALLY9','PVTALLY10'],value_name="vote_count")
# drop column "variable" created by .melt
tn_2016_pvtallymelt=tn_2016_pvtallymelt.drop(['variable'],axis=1)
tn_2016_pvtallymelt.head(9405)
```

```python
# add "party" column to tn_2016_namemelt
party=tn_2016_partymelt['party']
tn_2016_namemelt=tn_2016_namemelt.join(party)
tn_2016_namemelt.head(3)
```

```python
# add "vote_count" column to tn_2016_namemelt
pvtally=tn_2016_pvtallymelt['vote_count']
tn_2016_namemelt=tn_2016_namemelt.join(pvtally)
tn_2016_namemelt.head(3)
```

```python
tn_2016_namemelt.info()
```

```python
# lower case column names
tn_2016_namemelt.columns=tn_2016_namemelt.columns.str.lower()
tn_2016_namemelt.head(2)
```

```python
# change null values in "party" column to "no party"
tn_2016_namemelt['party'].fillna("No party", inplace = True)
tn_2016_namemelt.head(2)
```

```python
tn_2016_namemelt.info()
```

```python
# change null values in "vote_count" column to "0"
tn_2016_namemelt['vote_count'].fillna(0)
```

```python
tn_2016_namemelt.info()
```

### Looks safe to drop rows with no votes tallied

```python
# drop rows with no name and vote count
tn_2016_final=tn_2016_namemelt.dropna()
```

```python
# convert vote_count data from float64 to int 32 to to drop ".0"
tn_2016_final=tn_2016_final.astype({'vote_count':'int32'})
tn_2016_final
```

```python
# create data frame tn_2016_pres for presidental election
tn_2016_pres=tn_2016_final[tn_2016_final['officename'].str.match('United States President')]
tn_2016_pres=tn_2016_pres.reset_index()
tn_2016_pres
```

```python
tn_2016_pres['vote_count'].value_counts()
```

### 16240 rows have a vote count of 0. Drop those rows

```python
tn_2016_pres=tn_2016_pres[tn_2016_pres['vote_count'] != 0]
tn_2016_pres=tn_2016_pres.reset_index()
```

```python
# group county precincts into county per candidate 
tn_2016_pres=tn_2016_pres.groupby(['officename','county','name','party'], as_index=False)['vote_count'].sum()
tn_2016_pres
```

```python
tn_2016_pres.head(20)
```

### Add column that has total vote count by count for use in calculating county vote percentages

```python
tn_2016_pres['sum_county_vote']=tn_2016_pres.groupby(['vote_count'])['vote_count'].sum
tn_2016_pres
```

```python
filter=tn_2016_pres['county']==tn_2016_pres['county']
tn_2016_pres['sum_county_vote']=tn_2016_pres.where(filter).sum
tn_2016_pres
```

```python
# create new dataframe "county_vote" to use to find sum of votes per county
county_vote=tn_2016_pres.loc[0:,['county','vote_count']]
county_vote
```

```python
# sum votes per county
county_vote_sum=county_vote.groupby(['county'])['vote_count'].sum()
county_vote_sum
```

```python
# use "county_vote_sum" to add column "sum_county_vote" to tn_2016_pres correlating with value in each row
tn_2016_pres['sum_county_vote']=tn_2016_pres['county'].map(county_vote_sum)
tn_2016_pres.head(20)
```

```python
# create "percent_county_vote" column for tn_2016_pres
tn_2016_pres['percent_county_vote']=round(tn_2016_pres['vote_count']/tn_2016_pres['sum_county_vote']*100,2)
tn_2016_pres
```
