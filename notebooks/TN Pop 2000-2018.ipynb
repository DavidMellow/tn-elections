```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file TN_co-est2000-09-int-01-47.xls
tn_pop_2000=pd.read_excel('../data/US Census/TN_co-est2000-09-int-01-47.xls',
                         header=3)
tn_pop_2000
```

```python
tn_pop_2000.tail(10)
```

```python
tn_pop_2000.rename(columns={'Unnamed: 0':'geographic_area','Unnamed: 1':'april_1_2000','Unnamed: 12':'april_1_2010',
                   'Unnamed: 13':'july_1_2010'},inplace=True)
tn_pop_2000.head(2)
```

```python
# drop rows 96-103  
tn_pop_2000=tn_pop_2000[0:96]
tn_pop_2000.tail(10)
```

```python
tn_pop_2000.info()
```

```python
# clean up county names
tn_pop_2000.loc[:,'geographic_area']=tn_pop_2000['geographic_area'].str.replace('.','').str.replace(' ','').str.replace('County','')
tn_pop_2000.head(5)
```

```python
# only need election year columns (even years)
tn_pop_2000=tn_pop_2000.drop(['april_1_2000',2001,2003,2005,2007,2009,'april_1_2010','july_1_2010'],axis=1)
tn_pop_2000.head(5)
```

```python
# import excel file TN_co-est2010-19-annres-47.xlsx
tn_pop_2010=pd.read_excel('../data/US Census/TN_co-est2010-19-annres-47.xlsx',
                         header=3)
tn_pop_2010
```

```python
tn_pop_2010.tail(10)
```

```python
tn_pop_2010.rename(columns={'Unnamed: 0':'geographic_area'},inplace=True)
tn_pop_2010.head(2)
```

```python
# drop rows 96-103  df[:3] #keep top 3
tn_pop_2010=tn_pop_2010[0:96]
tn_pop_2010.tail(10)
```

```python
tn_pop_2010.info()
```

```python
# split "geographic_area" column to put "Tennesee" in new column   df[['First','Last']] = df.Name.str.split(expand=True) 
tn_pop_2010[['geographic_area','state']]=tn_pop_2010['geographic_area'].str.split(',',expand=True)
tn_pop_2010
```

```python
# clean up county names
tn_pop_2010.loc[:,'geographic_area']=tn_pop_2010['geographic_area'].str.replace('.','').str.replace(' ','').str.replace('County','')
tn_pop_2010.head(5)
```

```python
# only need election year columns (even years)
tn_pop_2010=tn_pop_2010.drop(['Census','Estimates Base',2011,2013,2015,2017,2019,'state'],axis=1)
tn_pop_2010.head(5)
```

```python
# merge dataframes into "tn_pop_2000_2018"
tn_pop_2000_2018=pd.merge(tn_pop_2000,tn_pop_2010, on='geographic_area')
tn_pop_2000_2018
```

```python
tn_pop_2000_2018.info()
```

```python
# change float64's to int32's to drop decimal
tn_pop_2000_2018=tn_pop_2000_2018.astype({2000:'int32',2002:'int32',2004:'int32',2006:'int32',2008:'int32',
                         2010:'int32',2012:'int32',2014:'int32',2016:'int32',2018:'int32',})
tn_pop_2000_2018.info()
```

```python
# add county index number for possible mapping
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# add "Tennessee" and "0" to first row to allow join
county_index.loc[-1]=['Tennessee',0]
county_index.index=county_index.index+1
county_index=county_index.sort_index()
county_index
```

```python
# join to df "tn_pop_2000_2018"
county_num=county_index['INDEX_NO']
tn_pop_2000_2018=tn_pop_2000_2018.join(county_num)
tn_pop_2000_2018
```

```python
tn_pop_2000_2018.to_excel(r'../data/US Census/tn_pop_2000_2018.xlsx',index=False)
```
