```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# import excel file November2008.xlsx
tn_2008=pd.read_excel('../data/2008_election/November2008.xlsx',
                      usecols=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE','BNAME1','TALLY1','BNAME2','TALLY2',
                               'BNAME3','TALLY3','BNAME4','TALLY4','BNAME5','TALLY5','BNAME6','TALLY6','BNAME7','TALLY7',
                               'BNAME8','TALLY8','BNAME9','TALLY9','BNAME10','TALLY10'])
tn_2008
```

```python
tn_2008.info()
```

```python
# merge entries in all RNAME columns into new column "name"
tn_2008_namemelt=tn_2008.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['BNAME1','BNAME2','BNAME3','BNAME4','BNAME5','BNAME6','BNAME7','BNAME8','BNAME9','BNAME10'],value_name="candidate")
# drop column "variable" created by .melt
tn_2008_namemelt=tn_2008_namemelt.drop(['variable'],axis=1)                          
tn_2008_namemelt.head()
```

```python
# merge entries in all TALLY columns into new column "vote_count"
tn_2008_pvtallymelt=tn_2008.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['TALLY1','TALLY2','TALLY3','TALLY4','TALLY5','TALLY6','TALLY7','TALLY8','TALLY9','TALLY10'],value_name="vote_count")
# drop column "variable" created by .melt
tn_2008_pvtallymelt=tn_2008_pvtallymelt.drop(['variable'],axis=1)
tn_2008_pvtallymelt.head()
```

```python
# add "vote_count" to tn_2008_namemelt
tally=tn_2008_pvtallymelt['vote_count']
tn_2008_namemelt=tn_2008_namemelt.join(tally)
tn_2008_namemelt.head(3)
```

```python
tn_2008_namemelt.info()
```

```python
# lower case column names
tn_2008_namemelt.columns=tn_2008_namemelt.columns.str.lower()
tn_2008_namemelt.head(2)
```

```python
tn_2008_namemelt.info()
```

### Look at cadidates with 0 votes, drop

```python
tn_2008_namemelt_0votes=tn_2008_namemelt[tn_2008_namemelt['vote_count'] == 0]
tn_2008_namemelt_0votes
```

```python
# drop rows with 0 in "vote_count" column
tn_2008_namemelt=tn_2008_namemelt[tn_2008_namemelt['vote_count'] != 0].reset_index(drop=True)
tn_2008_namemelt
```

```python
tn_2008_namemelt['vote_count'].value_counts()
```

```python
# group county precincts into county per candidate, in descending order of votes within county
tn_2008_int=tn_2008_namemelt.groupby(['electdate','officename','county','candidate'], as_index=False)['vote_count'].sum()
tn_2008_int=tn_2008_int.sort_values(by=['county','vote_count'],ascending=False).reset_index(drop=True)
tn_2008_int.head(30)
```

```python
# Capitalize couty names
tn_2008_int['county']=tn_2008_int['county'].str.title()
tn_2008_int.head(5)
```

```python
tn_2008_int.info()
```

```python
tn_2008_int['officename'].unique()
```

### Clean up names and greate separate party column

```python
# new dataframe new_names of non-write-in candidates
new_names=tn_2008_int[tn_2008_int['candidate'].str.contains("Write")==False]
new_names
```

```python
# split name fromm party desination
new_names=new_names['candidate'].str.split(pat='-',expand=True)
new_names
```

```python
new_names.rename(columns={0:'name',1:'party'},inplace=True)
new_names=new_names.drop([2],axis=1)
new_names
```

```python
# new dataframe tn_2012_int_wi of only write-in candidates
tn_2008_int_wi=tn_2008_int[tn_2008_int['candidate'].str.contains("\(")==False]
tn_2008_int_wi
```

```python
# split name fromm 'Write-in'
new_wi_name=tn_2008_int_wi['candidate'].str.split(pat='-',n=2,expand=True)
new_wi_name
```

```python
# create new 'party' columnn with "Write-In"
new_wi_name['party']=new_wi_name[[0,1]].agg('-'.join,axis=1)
new_wi_name
```

```python
#drop columns 0 and 1, rename column 2 to"candidate"
new_wi_name=new_wi_name.drop([0,1],axis=1)
new_wi_name
```

```python
new_wi_name.rename(columns={2:'name'},inplace=True)
new_wi_name
```

```python
# join new_names and new_wi_names
frames=[new_names,new_wi_name]
cleaned_names=pd.concat(frames)
cleaned_names
```

### Join cleaned names and party column to tn_2008_int

```python
name=cleaned_names['name']
party=cleaned_names['party']
tn_2008_final=tn_2008_int.join(name).join(party)
tn_2008_final
```

```python
tn_2008_final['party'].value_counts()
```

```python
# change "(R)" to "Republican" and "(D)"to "Democratic" in "party" column
tn_2008_final["party"]=tn_2008_final['party'].str.replace('\(R\)',"Republican").str.replace('\(D\)',"Democratic")
tn_2008_final
```

```python
# create data frame tn_2012_uss for senate election
tn_2008_uss=tn_2008_final[tn_2008_final['officename'].str.match('United States Senate')]
tn_2008_uss=tn_2008_uss.reset_index(drop=True)
tn_2008_uss
```

```python
tn_2008_uss.head(20)
```

### Add column that has total vote count by county for use in calculating county vote percentages

```python
# create new dataframe "county_vote" to use to find sum of votes per county
county_vote=tn_2008_uss.loc[0:,['county','vote_count']]
county_vote
```

```python
#sum county votes
county_vote_sum=county_vote.groupby(['county'])['vote_count'].sum()
county_vote_sum
```

```python
# use "county_vote_sum" to add column "sum_county_vote" to tn_2016_pres correlating with value in each row
tn_2008_uss['sum_county_vote']=tn_2008_uss['county'].map(county_vote_sum)
tn_2008_uss.head(20)
```

```python
# create "pct_county_vote" column for tn_2016_pres
tn_2008_uss['pct_county_vote']=round(tn_2008_uss['vote_count']/tn_2008_uss['sum_county_vote']*100,2)
tn_2008_uss.head(20)
```

```python
# how many parties?
tn_2008_uss['party'].value_counts()
```

## Create dataframe "tn_2008_pct_county"
_pct"  with percent values per party per county

```python
# create new dataframe with reduced columns values only
tn_2008_pct_county=tn_2008_uss.loc[0:,['electdate','county','party','pct_county_vote']]
tn_2008_pct_county
```

```python
# create new dataframe with Republican values only
tn_2008_rep_pct=tn_2008_pct_county[tn_2008_pct_county['party']== " Republican"].reset_index(drop=True)
tn_2008_rep_pct
```

```python
# rename 'pct_county_vote' to 'rep_pct' and drop "party" column to prepare for merge 
tn_2008_rep_pct=tn_2008_rep_pct.rename(columns={'pct_county_vote':'Republican'},inplace=False).drop(['party'], axis=1)
tn_2008_rep_pct
```

```python
# create new dataframe with Democratic values only
tn_2008_dem_pct=tn_2008_pct_county[tn_2008_pct_county['party']== " Democratic"].reset_index(drop=True)
tn_2008_dem_pct
```

```python
# rename 'pct_county_vote' to 'dem_pct' and drop "party" to prepare for merge 
tn_2008_dem_pct=tn_2008_dem_pct.rename(columns={'pct_county_vote':'Democratic'},inplace=False).drop(['party'], axis=1)
tn_2008_dem_pct
```

```python
# create new dataframe with non Republican or Democratic values only
tn_2008_other_pct=tn_2008_pct_county[(tn_2008_pct_county['party']!=" Republican")&(tn_2008_pct_county['party']!=" Democratic")].reset_index(drop=True)
tn_2008_other_pct
```

```python
# sum by county
tn_2008_other_pct=tn_2008_other_pct.groupby(['county'],as_index=False)['pct_county_vote'].sum()
tn_2008_other_pct
```

```python
# rename 'pct_county_vote' to 'other_pct' and drop "party" to prepare for merge 
tn_2008_other_pct=tn_2008_other_pct.rename(columns={'pct_county_vote':'Other'})
tn_2008_other_pct
```

```python
# join tn_2012_rep, tn_2012_dem_pct and tn_2012_other_pct to form tn_2012_county_uss_pct
dem=tn_2008_dem_pct['Democratic']
other=tn_2008_other_pct['Other']
tn_2008_county_uss_pct=tn_2008_rep_pct.join(dem).join(other)
tn_2008_county_uss_pct
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2008_county_uss_pct['Rep_Margin']= tn_2008_county_uss_pct['Republican']-tn_2008_county_uss_pct['Democratic']
tn_2008_county_uss_pct.head(50)
```

```python
# correct county names: Dekalb to DeKalb, Mcminn to McMinn, Mcnairy to McNairy 
tn_2008_county_uss_pct['county']=tn_2008_county_uss_pct['county'].str.replace('Dekalb','DeKalb').str.replace('Mcminn','McMinn').str.replace('Mcnairy','McNairy')
tn_2008_county_uss_pct.head(50)
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# join with tn_2012_county_uss_pct
tn_2008_county_uss_pct=tn_2008_county_uss_pct.merge(county_index,on="county")
tn_2008_county_uss_pct
```
