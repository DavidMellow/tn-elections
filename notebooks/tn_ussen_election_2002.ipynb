```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file 2002ussenate.xlsx
tn_2002=pd.read_excel('../data/2002_election/2002ussenate.xlsx')
tn_2002
```

```python
tn_2002.info()
```

```python
# drop columns "Unnamed: 11"
tn_2002=tn_2002.drop(columns=['Unnamed: 11'],axis=1)                     
tn_2002
```

```python
# merge columns with candidate name headers into new column "candidate" -- set aside for possible later use 
tn_2002_set_aside=tn_2002.melt(id_vars=['Electdate','County'],value_vars=['Clement, Bob - (D)','Alexander, Lamar - (R)','Baker, Wesley M. - (I)','Davidson, Karl Stanley - (I)','Gammon, Connie - (I)','Hooker, John Jay - (I)','Keplinger, H. Gary - (I)','Marceaux I, Basil J. - (I)','Write-Ins'],value_name='vote_count',var_name="candidate")                         
tn_2002_set_aside
```

```python
# create dataframe to use with vote counts by party by county
tn_2002CP=tn_2002.melt(id_vars=['Electdate','County','Clement, Bob - (D)','Alexander, Lamar - (R)'],value_vars=['Baker, Wesley M. - (I)','Davidson, Karl Stanley - (I)','Gammon, Connie - (I)','Hooker, John Jay - (I)','Keplinger, H. Gary - (I)','Marceaux I, Basil J. - (I)','Write-Ins'],value_name='Other')                         
# drop column "variable" created by .melt
tn_2002CP=tn_2002CP.drop(['variable'],axis=1) 
tn_2002CP
```

```python
# rename column "John F. Kerry - (D)" to "Democratic", "George W. Bush - (R)" to "Republican"
tn_2002CP.rename(columns={"Clement, Bob - (D)":"Democratic","Alexander, Lamar - (R)":"Republican"},inplace=True) 
tn_2002CP
```

```python
tn_2002CP.info()
```

### Look for 0 values in "Other" column indicating no votes

```python
tn_2002CP["Other"].value_counts()
```

###  61 are 0, drop those rows

```python
tn_2002CP=tn_2002CP[tn_2002CP["Other"]!= 0].reset_index(drop=True)
tn_2002CP
```

### Sever column "Other" so it can be totalled by county and then re-attached. Remaining df  should have 95 rows (one for each county)

```python
# create new df "tn_2002_other"
tn_2002_other=tn_2002CP.drop(columns=['Electdate','Democratic',"Republican"])
tn_2002_other
```

```python
# sum values in "Other" coumn by county
tn_2002_other=tn_2002_other.groupby(['County'],as_index=False)['Other'].sum()
tn_2002_other
```

```python
# see if dropping "Other" column of df tn_2004CP leaves 95 rows
tn_2002CP=tn_2002CP.drop(columns=['Other'])
tn_2002CP
```

#### It does not, drop duplicate rows

```python
tn_2002=tn_2002CP.drop_duplicates()
tn_2002
```

```python
# merge tn_2002 and tn_2002_other
other=tn_2002_other['Other']
tn_2002_uss=tn_2002.join(other)
tn_2002_uss
```

```python
tn_2002_uss.info()
```

### Add column that has total vote count by county for use in calculating county vote percentages

```python
# sum of votes per county
tn_2002_uss['sum_county_vote']=tn_2002_uss.sum(axis=1)
tn_2002_uss
```

```python
# create "pct_county_vote" column for tn_2004_pres
tn_2002_uss['Democratic_pct']=round(tn_2002_uss['Democratic']/tn_2002_uss['sum_county_vote']*100,2)
tn_2002_uss['Republican_pct']=round(tn_2002_uss['Republican']/tn_2002_uss['sum_county_vote']*100,2)
tn_2002_uss['Other_pct']=round(tn_2002_uss['Other']/tn_2002_uss['sum_county_vote']*100,2)
tn_2002_uss.head(20)
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2002_uss['Rep_Margin']= tn_2002_uss['Republican_pct']-tn_2002_uss['Democratic_pct']
tn_2002_uss.head(50)
```

```python
# Title county names in prepartion for merge with shape file
tn_2002_uss['County']=tn_2002_uss['County'].str.title()
tn_2002_uss
```

```python
# correct county names: Dekalb to DeKalb, Mcminn to McMinn, Mcnairy to McNairy 
tn_2002_uss['County']=tn_2002_uss['County'].str.replace('Dekalb','DeKalb').str.replace('Mcminn','McMinn').str.replace('Mcnairy','McNairy')
tn_2002_uss.tail(40)
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# join with tn_2002_uss
tn_2002_uss=tn_2002_uss.merge(county_index,left_on='County',right_on='county')
tn_2002_uss
```

```python
# drop duplicate "county" column
tn_2002_uss=tn_2002_uss.drop(columns=['county'])
tn_2002_uss
```
