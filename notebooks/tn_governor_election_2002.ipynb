```python
# drop "Unnamed: 5" and "Unnamed 9" columns which are all nulls
tn_2002g2=tn_2002g2.drop(columns=['Unnamed: 5','Unnamed: 9'],axis=1)
tn_2002g2.head(2)             
```

```python
# join dataframes tn_2002g1 and tn_2002g into tn_2002g
tn_2002g=tn_2002g1.merge(tn_2002g2,on='County')
tn_2002g
```

```python
# rename column "Electdate_x" to "Electdate"
tn_2002g.rename(columns={'Electdate_x':'Electdate'},inplace=True)
tn_2002g.head(2)
```

```python
# drop column "Electdate_y"
tn_2002g=tn_2002g.drop(columns=['Electdate_y'],axis=1)
tn_2002g.head(2)
```

```python
# merge columns with candidate name headers into new column "candidate" -- set aside for possible later use 
tn_2002g_set_aside=tn_2002g.melt(id_vars=['Electdate','County'],value_vars=['Bredesen, Phil - (D)','Hilleary, Van - (R)','Gatchell, David - (I)','Givens, Gabriel - (I)','Herren, James E. - (I)','Hooker, John Jay - (I)','Ledford, Ray - (I)','Leinoff, Marivuana Stout - (I)','Marceaux I, Basil J. - (I)','Sanders, Edwin C. - (I)','Simmons, Ronny - (I)','Waldron, Francis E. - (I)','Watson, Robert O. - (I)','Whitaker, Carl Two Feathers - (I)','Wilhoit, Jr., Charles V. - (I)','Write-Ins'],value_name='vote_count',var_name="candidate")                         
tn_2002g_set_aside
```

```python
# create dataframe to use with vote counts by party by county
tn_2002CP=tn_2002g.melt(id_vars=['Electdate','County','Bredesen, Phil - (D)','Hilleary, Van - (R)'],value_vars=['Gatchell, David - (I)','Givens, Gabriel - (I)','Herren, James E. - (I)','Hooker, John Jay - (I)','Ledford, Ray - (I)','Leinoff, Marivuana Stout - (I)','Marceaux I, Basil J. - (I)','Sanders, Edwin C. - (I)','Simmons, Ronny - (I)','Waldron, Francis E. - (I)','Watson, Robert O. - (I)','Whitaker, Carl Two Feathers - (I)','Wilhoit, Jr., Charles V. - (I)','Write-Ins'],value_name='Other')                         
# drop column "variable" created by .melt
tn_2002CP=tn_2002CP.drop(['variable'],axis=1) 
tn_2002CP
```

```python
# rename column "Phil Bredesen - D" to "Democratic", "Jim Bryson - R" to "Republican"
tn_2002CP.rename(columns={"Bredesen, Phil - (D)":"Democratic","Hilleary, Van - (R)":"Republican"},inplace=True) 
tn_2002CP
```

```python
tn_2002CP.info()
```

### Look for 0 values in "Other" column indicating no votes

```python
tn_2002CP["Other"].value_counts()
```

### 136 are 0, drop those rows

```python
tn_2002CP=tn_2002CP[tn_2002CP["Other"]!= 0].reset_index(drop=True)
tn_2002CP
```

### Sever column "Other" so it can be totalled by county and then re-attached. Remaining df  should have 95 rows (one for each county)

```python
# create new df "tn_2002_other"
tn_2002_other=tn_2002CP.drop(columns=['Electdate','Democratic',"Republican"])
tn_2002_other
```

```python
# sum values in "Other" coumn by county
tn_2002_other=tn_2002_other.groupby(['County'],as_index=False)['Other'].sum()
tn_2002_other
```

```python
# see if dropping "Other" column of df tn_2002CP leaves 95 rows
tn_2002CP=tn_2002CP.drop(columns=['Other'])
tn_2002CP
```

#### It does not, drop duplicate rows

```python
tn_2002_gov=tn_2002CP.drop_duplicates()
tn_2002_gov
```

```python
# merge tn_2002 pres and tn_2002_pres_other
other=tn_2002_other['Other']
tn_2002_gov=tn_2002_gov.join(other)
tn_2002_gov
```

```python
tn_2002_gov.info()
```

### Add column that has total vote count by county for use in calculating county vote percentages

```python
# sum of votes per county
tn_2002_gov['sum_county_vote']=tn_2002_gov.sum(axis=1)
tn_2002_gov
```

```python
# create "pct_county_vote" column for tn_2002_gov
tn_2002_gov['Democratic_pct']=round(tn_2002_gov['Democratic']/tn_2002_gov['sum_county_vote']*100,2)
tn_2002_gov['Republican_pct']=round(tn_2002_gov['Republican']/tn_2002_gov['sum_county_vote']*100,2)
tn_2002_gov['Other_pct']=round(tn_2002_gov['Other']/tn_2002_gov['sum_county_vote']*100,2)
tn_2002_gov.head(20)
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2002_gov['Rep_Margin']= tn_2002_gov['Republican_pct']-tn_2002_gov['Democratic_pct']
tn_2002_gov.head(50)
```

```python
# Title county names in prepartion for merge with shape file
tn_2002_gov['County']=tn_2002_gov['County'].str.title()
tn_2002_gov
```

```python
# correct county names: Dekalb to DeKalb, Mcminn to McMinn, Mcnairy to McNairy 
tn_2002_gov['County']=tn_2002_gov['County'].str.replace('Dekalb','DeKalb').str.replace('Mcminn','McMinn').str.replace('Mcnairy','McNairy')
tn_2002_gov.tail(40)
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# join with tn_2002_gov
tn_2002_gov=tn_2002_gov.merge(county_index,left_on='County',right_on='county')
tn_2002_gov
```

```python
# drop duplicate "county" column
tn_2002_gov=tn_2002_gov.drop(columns=['county'])
tn_2002_gov
```
