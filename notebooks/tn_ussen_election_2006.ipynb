```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file 2006ussenate1.xlsx
tn_2006uss1=pd.read_excel('../data/2006_election/2006ussenate1.xlsx')
tn_2006uss1
```

```python
tn_2006uss1.info()
```

```python
# drop "Unnamed 12" which is all nulls
tn_2006uss1=tn_2006uss1.drop(columns=['Unnamed: 12'],axis=1)
tn_2006uss1.head(2)                         
```

```python
# import excel file 2006ussenate2.xlsx
tn_2006uss2=pd.read_excel('../data/2006_election/2006ussenate2.xlsx')
tn_2006uss2
```

```python
tn_2006uss2.info()
```

```python
# drop "Unnamed: 3" which is all nulls
tn_2006uss2=tn_2006uss2.drop(columns=['Unnamed: 3'],axis=1)
tn_2006uss2.head(2)             
```

```python
# join dataframes tn_2006uss1 and tn_2006uss2 into tn_2002uss
tn_2006uss=tn_2006uss1.merge(tn_2006uss2,on='County')
tn_2006uss
```

```python
tn_2006uss.info()
```

```python
# merge columns with candidate name headers into new column "candidate" -- set aside for possible later use 
tn_2006uss_set_aside=tn_2006uss.melt(id_vars=['Electdate','County'],value_vars=['Harold Ford, Jr. - D','Bob Corker - R','Ed Choate - I','David Gatchell - I','Emory "Bo" Heyward - I','Gary Keplinger - I','Christopher Joseph Lugo - I','Write-in - Thomas M. Davidson - I','Write-in - James Anthony Gray - I','Write-in - John Jay Hooker - I','Write-in - Gloria D. Reagon Price - I','Write-in - Mary Taylor-Shelby - I'],value_name='vote_count',var_name="candidate")                         
tn_2006uss_set_aside
```

```python
# create dataframe to use with vote counts by party by county
tn_2006CP=tn_2006uss.melt(id_vars=['Electdate','County','Harold Ford, Jr. - D','Bob Corker - R'],value_vars=['Ed Choate - I','David Gatchell - I','Emory "Bo" Heyward - I','Gary Keplinger - I','Christopher Joseph Lugo - I','Write-in - Thomas M. Davidson - I','Write-in - James Anthony Gray - I','Write-in - John Jay Hooker - I','Write-in - Gloria D. Reagon Price - I','Write-in - Mary Taylor-Shelby - I'],value_name='Other')                         
# drop column "variable" created by .melt
tn_2006CP=tn_2006CP.drop(['variable'],axis=1) 
tn_2006CP
```

```python
# rename column "Phil Bredesen - D" to "Democratic", "Jim Bryson - R" to "Republican"
tn_2006CP.rename(columns={"Harold Ford, Jr. - D":"Democratic","Bob Corker - R":"Republican"},inplace=True) 
tn_2006CP
```

```python
tn_2006CP.info()
```

### Look for 0 values in "Other" column indicating no votes

```python
tn_2006CP["Other"].value_counts()
```

### 464 are 0, drop those rows

```python
tn_2006CP=tn_2006CP[tn_2006CP["Other"]!= 0].reset_index(drop=True)
tn_2006CP
```

### Sever column "Other" so it can be totalled by county and then re-attached. Remaining df  should have 95 rows (one for each county)

```python
# create new df "tn_2006_other"
tn_2006_other=tn_2006CP.drop(columns=['Electdate','Democratic',"Republican"])
tn_2006_other
```

```python
# sum values in "Other" coumn by county
tn_2006_other=tn_2006_other.groupby(['County'],as_index=False)['Other'].sum()
tn_2006_other
```

```python
# see if dropping "Other" column of df tn_2002CP leaves 95 rows
tn_2006=tn_2006CP.drop(columns=['Other'])
tn_2006
```

#### It does not, drop duplicate rows

```python
tn_2006=tn_2006.drop_duplicates()
tn_2006
```

```python
# merge tn_2006 and tn_2006_other
other=tn_2006_other['Other']
tn_2006_uss=tn_2006.join(other)
tn_2006_uss
```

```python
tn_2006_uss.info()
```

### Add column that has total vote count by county for use in calculating county vote percentages

```python
# sum of votes per county
tn_2006_uss['sum_county_vote']=tn_2006_uss.sum(axis=1)
tn_2006_uss
```

```python
# create "pct_county_vote" column for tn_2006_uss
tn_2006_uss['Democratic_pct']=round(tn_2006_uss['Democratic']/tn_2006_uss['sum_county_vote']*100,2)
tn_2006_uss['Republican_pct']=round(tn_2006_uss['Republican']/tn_2006_uss['sum_county_vote']*100,2)
tn_2006_uss['Other_pct']=round(tn_2006_uss['Other']/tn_2006_uss['sum_county_vote']*100,2)
tn_2006_uss.head(20)
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2006_uss['Rep_Margin']= tn_2006_uss['Republican_pct']-tn_2006_uss['Democratic_pct']
tn_2006_uss.head(50)
```

```python
# Title county names in prepartion for merge with shape file
tn_2006_uss['County']=tn_2006_uss['County'].str.title()
tn_2006_uss
```

```python
# correct county names: Dekalb to DeKalb, Mcminn to McMinn, Mcnairy to McNairy 
tn_2006_uss['County']=tn_2006_uss['County'].str.replace('Dekalb','DeKalb').str.replace('Mcminn','McMinn').str.replace('Mcnairy','McNairy')
tn_2006_uss.tail(40)
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# join with tn_2006_uss
tn_2006_uss=tn_2006_uss.merge(county_index,left_on='County',right_on='county')
tn_2006_uss
```

```python
# drop duplicate "county" column
tn_2006_uss=tn_2006_uss.drop(columns=['county'])
tn_2006_uss
```
