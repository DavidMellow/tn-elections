```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file 2006Gov.xlsx
tn_2006g=pd.read_excel('../data/2006_election/2006Gov.xlsx')
tn_2006g
```

```python
tn_2006g.info()
```

```python
# drop columns "Unnamed: 11"
tn_2006g=tn_2006g.drop(columns=['Unnamed: 11'],axis=1)                     
tn_2006g
```

```python
# merge columns with candidate name headers into new column "candidate" -- set aside for possible later use 
tn_2006g_set_aside=tn_2006g.melt(id_vars=['Electdate','County'],value_vars=['Phil Bredesen - D','Jim Bryson - R','George Banks - I','David Gatchell - I','Marivuana Stout Leinoff - I','Charles E. Smith - I','Howard M. Switzer - I','Carl "Twofeathers" Whitaker - I','Write-in - John Jay Hooker - I'],value_name='vote_count',var_name="candidate")                         
tn_2006g_set_aside
```

```python
# create dataframe to use with vote counts by party by county
tn_2006CP=tn_2006g.melt(id_vars=['Electdate','County','Phil Bredesen - D','Jim Bryson - R'],value_vars=['George Banks - I','David Gatchell - I','Marivuana Stout Leinoff - I','Charles E. Smith - I','Howard M. Switzer - I','Carl "Twofeathers" Whitaker - I','Write-in - John Jay Hooker - I'],value_name='Other')                         
# drop column "variable" created by .melt
tn_2006CP=tn_2006CP.drop(['variable'],axis=1) 
tn_2006CP
```

```python
# rename column "Phil Bredesen - D" to "Democratic", "Jim Bryson - R" to "Republican"
tn_2006CP.rename(columns={"Phil Bredesen - D":"Democratic","Jim Bryson - R":"Republican"},inplace=True) 
tn_2006CP
```

```python
tn_2006CP.info()
```

### Look for 0 values in "Other" column indicating no votes

```python
tn_2006CP["Other"].value_counts()
```

### 99 are 0, drop those rows

```python
tn_2006CP=tn_2006CP[tn_2006CP["Other"]!= 0].reset_index(drop=True)
tn_2006CP
```

### Sever column "Other" so it can be totalled by county and then re-attached. Remaining df  should have 95 rows (one for each county)

```python
# create new df "tn_2006_other"
tn_2006_other=tn_2006CP.drop(columns=['Electdate','Democratic',"Republican"])
tn_2006_other
```

```python
# sum values in "Other" coumn by county
tn_2006_pres_other=tn_2006_other.groupby(['County'],as_index=False)['Other'].sum()
tn_2006_pres_other
```

```python
# see if dropping "Other" column of df tn_2006CP leaves 95 rows
tn_2006CP=tn_2006CP.drop(columns=['Other'])
tn_2006CP
```

#### It does not, drop duplicate rows

```python
tn_2006_gov=tn_2006CP.drop_duplicates()
tn_2006_gov
```

```python
# merge tn_2006 pres and tn_2006_pres_other
other=tn_2006_other['Other']
tn_2006_gov=tn_2006_gov.join(other)
tn_2006_gov
```

```python
tn_2006_gov.info()
```

### Add column that has total vote count by county for use in calculating county vote percentages

```python
# sum of votes per county
tn_2006_gov['sum_county_vote']=tn_2006_gov.sum(axis=1)
tn_2006_gov
```

```python
# create "pct_county_vote" column for tn_2004_pres
tn_2006_gov['Democratic_pct']=round(tn_2006_gov['Democratic']/tn_2006_gov['sum_county_vote']*100,2)
tn_2006_gov['Republican_pct']=round(tn_2006_gov['Republican']/tn_2006_gov['sum_county_vote']*100,2)
tn_2006_gov['Other_pct']=round(tn_2006_gov['Other']/tn_2006_gov['sum_county_vote']*100,2)
tn_2006_gov.head(20)
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2006_gov['Rep_Margin']= tn_2006_gov['Republican_pct']-tn_2006_gov['Democratic_pct']
tn_2006_gov.head(50)
```

```python
# Title county names in prepartion for merge with shape file
tn_2006_gov['County']=tn_2006_gov['County'].str.title()
tn_2006_gov
```

```python
# correct county names: Dekalb to DeKalb, Mcminn to McMinn, Mcnairy to McNairy 
tn_2006_gov['County']=tn_2006_gov['County'].str.replace('Dekalb','DeKalb').str.replace('Mcminn','McMinn').str.replace('Mcnairy','McNairy')
tn_2006_gov.tail(40)
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# join with tn_2004_pres
tn_2006_gov=tn_2006_gov.merge(county_index,left_on='County',right_on='county')
tn_2006_gov
```

```python
# drop duplicate "county" column
tn_2006_gov=tn_2006_gov.drop(columns=['county'])
tn_2006_gov
```
