```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file November2004P.xlsx
tn_2004P=pd.read_excel('../data/2004_election/November2004P.xlsx')
tn_2004P
```

```python
tn_2004P.info()
```

```python
# drop columns "Unnamed: 9"
tn_2004P=tn_2004P.drop(columns=['Unnamed: 9'],axis=1)                     
tn_2004P
```

```python
# merge columns with candidate name headers into new column "candidate" -- set aside for possible later use 
tn_2004P_set_aside=tn_2004P.melt(id_vars=['Electdate','County'],value_vars=['John F. Kerry - (D)','George W. Bush - (R)','Michael Badnarik - (I)','Ralph Nader - (I)','Michael Anthony Peroutka - (I)','Walt Brown - (Write-In)','David Cobb - (Write-In)'],value_name='vote_count',var_name="candidate")                         
tn_2004P_set_aside
```

```python
# create dataframe to use with vote counts by party by county
tn_2004CP=tn_2004P.melt(id_vars=['Electdate','County','John F. Kerry - (D)','George W. Bush - (R)'],value_vars=['Michael Badnarik - (I)','Ralph Nader - (I)','Michael Anthony Peroutka - (I)','Walt Brown - (Write-In)','David Cobb - (Write-In)'],value_name='Other')                         
# drop column "variable" created by .melt
tn_2004CP=tn_2004CP.drop(['variable'],axis=1) 
tn_2004CP
```

```python
# rename column "John F. Kerry - (D)" to "Democratic", "George W. Bush - (R)" to "Republican"
tn_2004CP.rename(columns={"John F. Kerry - (D)":"Democratic","George W. Bush - (R)":"Republican"},inplace=True) 
tn_2004CP
```

```python
tn_2004CP.info()
```

### Look for 0 values in "Other" column indicating no votes

```python
tn_2004CP["Other"].value_counts()
```

### 177 are 0, drop those rows

```python
tn_2004CP=tn_2004CP[tn_2004CP["Other"]!= 0].reset_index(drop=True)
tn_2004CP
```

### Sever column "Other" so it can be totalled by county and then re-attached. Remaining df  should have 95 rows (one for each county)

```python
# create new df "tn_2004_other"
tn_2004_other=tn_2004CP.drop(columns=['Electdate','Democratic',"Republican"])
tn_2004_other
```

```python
# sum values in "Other" coumn by county
tn_2004_pres_other=tn_2004_other.groupby(['County'],as_index=False)['Other'].sum()
tn_2004_pres_other
```

```python
# see if dropping "Other" column of df tn_2004CP leaves 95 rows
tn_2004CP=tn_2004CP.drop(columns=['Other'])
tn_2004CP
```

#### It does not, drop duplicate rows

```python
tn_2004_pres=tn_2004CP.drop_duplicates()
tn_2004_pres
```

```python
# merge tn_2004 pres and tn_2004_pres_other
other=tn_2004_pres_other['Other']
tn_2004_pres=tn_2004_pres.join(other)
tn_2004_pres
```

```python
tn_2004_pres.info()
```

### Add column that has total vote count by county for use in calculating county vote percentages

```python
# sum of votes per county
tn_2004_pres['sum_county_vote']=tn_2004_pres.sum(axis=1)
tn_2004_pres
```

```python
# create "pct_county_vote" column for tn_2004_pres
tn_2004_pres['Democratic_pct']=round(tn_2004_pres['Democratic']/tn_2004_pres['sum_county_vote']*100,2)
tn_2004_pres['Republican_pct']=round(tn_2004_pres['Republican']/tn_2004_pres['sum_county_vote']*100,2)
tn_2004_pres['Other_pct']=round(tn_2004_pres['Other']/tn_2004_pres['sum_county_vote']*100,2)
tn_2004_pres.head(20)
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2004_pres['Rep_Margin']= tn_2004_pres['Republican_pct']-tn_2004_pres['Democratic_pct']
tn_2004_pres.head(50)
```

```python
# Title county names in prepartion for merge with shape file
tn_2004_pres['County']=tn_2004_pres['County'].str.title()
tn_2004_pres
```

```python
# correct county names: Dekalb to DeKalb, Mcminn to McMinn, Mcnairy to McNairy 
tn_2004_pres['County']=tn_2004_pres['County'].str.replace('Dekalb','DeKalb').str.replace('Mcminn','McMinn').str.replace('Mcnairy','McNairy')
tn_2004_pres.tail(40)
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# join with tn_2004_pres
tn_2004_pres=tn_2004_pres.merge(county_index,left_on='County',right_on='county')
tn_2004_pres
```

```python
# drop duplicate "county" column
tn_2004_pres=tn_2004_pres.drop(columns=['county'])
tn_2004_pres
```
