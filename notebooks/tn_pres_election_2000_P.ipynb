```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file November2000.xlsx
tn_2000P=pd.read_excel('../data/2000_election/November2000.xlsx',header=1)
tn_2000P
```

```python
tn_2000P.info()
```

```python
# drop columns "Unnamed: 4" and "Unnamed: 13" which are all null
tn_2000P=tn_2000P.drop(columns=['Unnamed: 4','Unnamed: 13'],axis=1)                     
tn_2000P
```

```python
# merge columns with candidate name headers into new column "candidate" -- set aside for possible later use 
tn_2000P_set_aside=tn_2000P.melt(id_vars=['Electdate','County'],value_vars=['Gore, Al - D','Bush, George W. - R','Brown, Cathy Gordon - I','Browne, Harry - I (Lib)','Buchanan, Patrick J. - I (Ref)','Hagelin, John - I (Ref)','Nader, Ralph - I (Green)','Phillips, Howard - I','Venson, Randall - I','Write-Ins'],value_name='vote_count',var_name="candidate")                         
tn_2000P_set_aside
```

```python
# create dataframe to use with vote counts by party by county
tn_2000CP=tn_2000P.melt(id_vars=['Electdate','County','Gore, Al - D','Bush, George W. - R'],value_vars=['Brown, Cathy Gordon - I','Browne, Harry - I (Lib)','Buchanan, Patrick J. - I (Ref)','Hagelin, John - I (Ref)','Nader, Ralph - I (Green)','Phillips, Howard - I','Venson, Randall - I','Write-Ins'],value_name='Other')                         
# drop column "variable" created by .melt
tn_2000CP=tn_2000CP.drop(['variable'],axis=1) 
tn_2000CP
```

```python
# rename column "Gore, Al - D" to "Democratic", "George W. Bush - (R)" to "Republican"
tn_2000CP.rename(columns={"Gore, Al - D":"Democratic","Bush, George W. - R":"Republican"},inplace=True) 
tn_2000CP
```

```python
tn_2000CP.info()
```

### Look for 0 values in "Other" column indicating no votes

```python
tn_2000CP["Other"].value_counts()
```

# 110 are 0, drop those rows

```python
tn_2000CP=tn_2000CP[tn_2000CP["Other"]!= 0].reset_index(drop=True)
tn_2000CP
```

### Sever column "Other" so it can be totalled by county and then re-attached. Remaining df  should have 95 rows (one for each county)

```python
# create new df "tn_2000_other"
tn_2000_other=tn_2000CP.drop(columns=['Electdate','Democratic',"Republican"])
tn_2000_other
```

```python
# sum values in "Other" coumn by county
tn_2000_pres_other=tn_2000_other.groupby(['County'],as_index=False)['Other'].sum()
tn_2000_pres_other
```

```python
# see if dropping "Other" column of df tn_2004CP leaves 95 rows
tn_2000CP=tn_2000CP.drop(columns=['Other'])
tn_2000CP
```

#### It does not, drop duplicate rows

```python
tn_2000_pres=tn_2000CP.drop_duplicates().reset_index(drop=True)
tn_2000_pres
```

```python
# merge tn_2000 pres and tn_2000_pres_other
other=tn_2000_pres_other['Other']
tn_2000_pres=tn_2000_pres.join(other)
tn_2000_pres
```

```python
tn_2000_pres.info()
```

### Add column that has total vote count by county for use in calculating county vote percentages

```python
# sum of votes per county
tn_2000_pres['sum_county_vote']=tn_2000_pres.sum(axis=1)
tn_2000_pres
```

```python
# create "pct_county_vote" column for tn_2004_pres
tn_2000_pres['Democratic_pct']=round(tn_2000_pres['Democratic']/tn_2000_pres['sum_county_vote']*100,2)
tn_2000_pres['Republican_pct']=round(tn_2000_pres['Republican']/tn_2000_pres['sum_county_vote']*100,2)
tn_2000_pres['Other_pct']=round(tn_2000_pres['Other']/tn_2000_pres['sum_county_vote']*100,2)
tn_2000_pres.head(20)
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2000_pres['Rep_Margin']= tn_2000_pres['Republican_pct']-tn_2000_pres['Democratic_pct']
tn_2000_pres.head(50)
```

```python
# Title county names in prepartion for merge with shape file
tn_2000_pres['County']=tn_2000_pres['County'].str.title()
tn_2000_pres
```

```python
# correct county names: Dekalb to DeKalb, Mcminn to McMinn, Mcnairy to McNairy 
tn_2000_pres['County']=tn_2000_pres['County'].str.replace('Dekalb','DeKalb').str.replace('Mcminn','McMinn').str.replace('Mcnairy','McNairy')
tn_2000_pres.tail(40)
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# join with tn_2000_pres
tn_2000_pres=tn_2000_pres.merge(county_index,left_on='County',right_on='county')
tn_2000_pres
```

```python
# drop duplicate "county" column
tn_2000_pres=tn_2000_pres.drop(columns=['county'])
tn_2000_pres
```
