```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# import excel file Nov 2018 General results.xlsx
tn_2018g=pd.read_excel('../data/2018_election/Nov 2018 General results.xlsx',
                      usecols=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE','RNAME1','PARTY1','PVTALLY1',
                             'RNAME2','PARTY2','PVTALLY2','RNAME3','PARTY3','PVTALLY3','RNAME4','PARTY4','PVTALLY4',
                             'RNAME5','PARTY5','PVTALLY5','RNAME6','PARTY6','PVTALLY6','RNAME7','PARTY7','PVTALLY7',
                             'RNAME8','PARTY8','PVTALLY8','RNAME9','PARTY9','PVTALLY9','RNAME10','PARTY10','PVTALLY10'])
tn_2018g
```

```python
tn_2018g.info()
```

```python
# merge entries in all RNAME columns into new column "name"
tn_2018g_namemelt=tn_2018g.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['RNAME1','RNAME2','RNAME3','RNAME4','RNAME5','RNAME6','RNAME7','RNAME8','RNAME9','RNAME10'],value_name="name")
# drop column "variable" created by .melt
tn_2018g_namemelt=tn_2018g_namemelt.drop(['variable'],axis=1)                          
tn_2018g_namemelt.head(9405)
```

```python
# merge entries in all PARTY columns into new column "party"
tn_2018g_partymelt=tn_2018g.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['PARTY1','PARTY2','PARTY3','PARTY4','PARTY5','PARTY6','PARTY7','PARTY8','PARTY9','PARTY10'],value_name="party")
# drop column "variable" created by .melt
tn_2018g_partymelt=tn_2018g_partymelt.drop(['variable'],axis=1)
tn_2018g_partymelt.head(9405)
```

```python
# merge entries in all PVTALLY columns into new column "vote_count"
tn_2018g_pvtallymelt=tn_2018g.melt(id_vars=['COUNTY','PRECINCT','OFFICENAME','ELECTDATE'],value_vars=['PVTALLY1','PVTALLY2','PVTALLY3','PVTALLY4','PVTALLY5','PVTALLY6','PVTALLY7','PVTALLY8','PVTALLY9','PVTALLY10'],value_name="vote_count")
# drop column "variable" created by .melt
tn_2018g_pvtallymelt=tn_2018g_pvtallymelt.drop(['variable'],axis=1)
tn_2018g_pvtallymelt.head(9405)
```

```python
# add "party" and "vote_count" columns to tn_2016_namemelt
party=tn_2018g_partymelt['party']
pvtally=tn_2018g_pvtallymelt['vote_count']
tn_2018g_namemelt=tn_2018g_namemelt.join(party).join(pvtally)
tn_2018g_namemelt.head(3)
```

```python
tn_2018g_namemelt.info()
```

```python
# lower case column names
tn_2018g_namemelt.columns=tn_2018g_namemelt.columns.str.lower()
tn_2018g_namemelt
```

```python
# change null values in "party" column to "no party"
tn_2018g_namemelt['party'].fillna("No party", inplace = True)
tn_2018g_namemelt
```

```python
tn_2018g_namemelt.info()
```

```python
# change null values in "vote_count" column to "0"
tn_2018g_namemelt['vote_count'].fillna(0)
```

### Add column that has total vote count by count for use in calculating county vote percentages

```python
# group county precincts into county per candidate, in descending order of votes within county
tn_2018_uss_int=tn_2018_uss_int.groupby(['electdate','officename','county','name','party'], as_index=False)['vote_count'].sum()
tn_2018_uss=tn_2018_uss_int.sort_values(by=['county','vote_count'],ascending=False).reset_index(drop=True)
tn_2018_uss.head(30)
```

```python
# create new dataframe "county_vote" to use to find sum of votes per county
county_vote=tn_2018_uss.loc[0:,['county','vote_count']]
county_vote
```

```python
#sum county votes
county_vote_sum=county_vote.groupby(['county'])['vote_count'].sum()
county_vote_sum
```

```python
# use "county_vote_sum" to add column "sum_county_vote" to tn_2018_uss correlating with value in each row
tn_2018_uss['sum_county_vote']=tn_2018_uss['county'].map(county_vote_sum)
tn_2018_uss.head(20)
```

```python
# create "pct_county_vote" column for tn_2018_uss
tn_2018_uss['pct_county_vote']=round(tn_2018_uss['vote_count']/tn_2018_uss['sum_county_vote']*100,2)
tn_2018_uss
```

```python
# how many parties?
tn_2018_uss['party'].value_counts()
```

### Create dataframe "tn_2018_uss_pct_county"  with percent values per party per county

```python
# create new dataframe with reduced columns values only
tn_2018_uss_pct_county=tn_2018_uss.loc[0:,['electdate','officename','county','party','pct_county_vote']]
tn_2018_uss_pct_county
```

```python
# create new dataframe with Republican values only
tn_2018_uss_rep_pct=tn_2018_uss_pct_county[tn_2018_uss_pct_county['party']== "Republican"].reset_index(drop=True)
tn_2018_uss_rep_pct
```

```python
# rename 'pct_county_vote' to 'rep_pct' and drop "party" column to prepare for merge 
tn_2018_uss_rep_pct=tn_2018_uss_rep_pct.rename(columns={'pct_county_vote':'Republican'},inplace=False).drop(['party'], axis=1)
tn_2018_uss_rep_pct
```

```python
# create new dataframe with Democratic values only
tn_2018_uss_dem_pct=tn_2018_uss_pct_county[tn_2018_uss_pct_county['party']== "Democratic"].reset_index(drop=True)
tn_2018_uss_dem_pct
```

```python
# rename 'pct_county_vote' to 'dem_pct' and drop "party" to prepare for merge 
tn_2018_uss_dem_pct=tn_2018_uss_dem_pct.rename(columns={'pct_county_vote':'Democratic'},inplace=False).drop(['party'], axis=1)
tn_2018_uss_dem_pct
```

```python
# create new dataframe with non Republican or Democratic values only
tn_2018_uss_other_pct=tn_2018_uss_pct_county[(tn_2018_uss_pct_county['party']!="Republican")&(tn_2018_uss_pct_county['party']!="Democratic")].reset_index(drop=True)
tn_2018_uss_other_pct
```

```python
# sum by county
tn_2018_uss_other_pct=tn_2018_uss_other_pct.groupby(['county'],as_index=False)['pct_county_vote'].sum()
tn_2018_uss_other_pct
```

```python
# rename 'pct_county_vote' to 'other_pct' and drop "party" to prepare for merge 
tn_2018_uss_other_pct=tn_2018_uss_other_pct.rename(columns={'pct_county_vote':'Other'})
tn_2018_uss_other_pct
```

```python
# join tn_2018_uss_rep, tn_2018_uss_dem_pct and tn_2018__other_pct to form tn_2018_uss_county_pct
dem=tn_2018_uss_dem_pct['Democratic']
other=tn_2018_uss_other_pct['Other']
tn_2018_uss_county_pct=tn_2018_uss_rep_pct.join(dem).join(other)
tn_2018_uss_county_pct
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2018_uss_county_pct['Rep_Margin']= tn_2018_uss_county_pct['Republican']-tn_2018_uss_county_pct['Democratic']
tn_2018_uss_county_pct
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```

```python
# join with tn_2018_uss_county_pct
tn_2018_uss_county_pct=tn_2018_uss_county_pct.merge(county_index,on="county")
tn_2018_uss_county_pct.head(50)
```
