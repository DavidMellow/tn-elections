```python
# make one big happy df "tn_2000_2018_to" (result = pd.concat([df1, df4], ignore_index=True, sort=False))
tn_2000_2018_to=pd.concat([tn_2018_to, tn_2016_to, tn_2014_to, tn_2012_to, tn_2010_to, tn_2008_to, tn_2006_to,
                           tn_2004_to, tn_2002_to, tn_2000_to], ignore_index=True, sort=False)
#tn_2000_2018_to=pd.merge(tn_2018_to, tn_2016_to,on="County")
#tn_2000_2018_to=tn_2018_to.append(tn_2016_to, sort=False)
tn_2000_2018_to
```

```python
tn_2000_2018_to.info()
```

```python
# drop column "Unnamed: 9"
tn_2000_2018_to=tn_2000_2018_to.drop(columns=['Unnamed: 9'])
tn_2000_2018_to.info()
```

```python
# convert decimal values to percents in columns 'Voter Turnout', 'Early Voters Pct' and 'Early and Absentee Voters Pct'
tn_2000_2018_to['Voter Turnout Pct']=tn_2000_2018_to['Voter Turnout Pct']*100
tn_2000_2018_to['Early Voters Pct']=tn_2000_2018_to['Early Voters Pct']*100
tn_2000_2018_to['Early and Absentee Voters Pct']=tn_2000_2018_to['Early and Absentee Voters Pct']*100
tn_2000_2018_to
```

```python
tn_2000_2018_to.info()
```

```python
# new dataframe of total values only
tn_2000_2018_to_totals=tn_2000_2018_to[tn_2000_2018_to['County']=='Total']
tn_2000_2018_to_totals
```

```python
# add column for in person voting
tn_2000_2018_to_totals['In Person Voters']=tn_2000_2018_to_totals['Total Votes Cast']-(tn_2000_2018_to_totals['Absentee By Mail Voters']+tn_2000_2018_to_totals['Early Voters'])
tn_2000_2018_to_totals
```

```python
# add percentage of in person voters of votes cast
tn_2000_2018_to_totals['In Person Voters Pct']=tn_2000_2018_to_totals['In Person Voters']/tn_2000_2018_to_totals['Total Votes Cast']*100
tn_2000_2018_to_totals
```

```python
tn_2000_2018_to_totals['In Person Voters Pct']=tn_2000_2018_to_totals['In Person Voters Pct'].round(2)
tn_2000_2018_to_totals
```

```python
tn_2000_2018_to_totals.to_excel(r'../data/TN_Voter_Turnout/turnout_xlsx/tn_2000_2018_to_totals.xlsx',index=False)
```

```python
#read TN_county_index_shape.csv to add index from TN Counties shape file for joining in Tableau
county_index=pd.read_csv('../data/TN_county_index_shape.csv')
county_index
```
