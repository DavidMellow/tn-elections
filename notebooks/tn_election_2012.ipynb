### Add column that has total vote count by count for use in calculating county vote percentages

```python
# create new dataframe "county_vote" to use to find sum of votes per county
county_vote=tn_2012_pres.loc[0:,['county','vote_count']]
county_vote
```

```python
#sum county votes
county_vote_sum=county_vote.groupby(['county'])['vote_count'].sum()
county_vote_sum
```

```python
# use "county_vote_sum" to add column "sum_county_vote" to tn_2016_pres correlating with value in each row
tn_2012_pres['sum_county_vote']=tn_2012_pres['county'].map(county_vote_sum)
tn_2012_pres.head(20)
```

```python
# create "pct_county_vote" column for tn_2016_pres
tn_2012_pres['pct_county_vote']=round(tn_2012_pres['vote_count']/tn_2012_pres['sum_county_vote']*100,2)
tn_2012_pres.head(20)
```

## Create dataframe "tn_2012_county_pres_pct"  with percent values per party per county

```python
# create new dataframe with reduced columns values only
tn_2012_pct_county=tn_2012_pres.loc[0:,['county','party','pct_county_vote']]
tn_2012_pct_county
```

```python
# create new dataframe with Republican values only
tn_2012_rep_pct=tn_2012_pct_county[tn_2012_pct_county['party']== " Republican"].reset_index(drop=True)
tn_2012_rep_pct
```

```python
# rename 'pct_county_vote' to 'rep_pct' and drop "party" column to prepare for merge 
tn_2012_rep_pct=tn_2012_rep_pct.rename(columns={'pct_county_vote':'Republican'},inplace=False).drop(['party'], axis=1)
tn_2012_rep_pct
```

```python
# create new dataframe with Democratic values only
tn_2012_dem_pct=tn_2012_pct_county[tn_2012_pct_county['party']== " Democratic"].reset_index(drop=True)
tn_2012_dem_pct
```

```python
# rename 'pct_county_vote' to 'dem_pct' and drop "party" to prepare for merge 
tn_2012_dem_pct=tn_2012_dem_pct.rename(columns={'pct_county_vote':'Democratic'},inplace=False).drop(['party'], axis=1)
tn_2012_dem_pct
```

```python
# create new dataframe with non Republican or Democratic values only
tn_2012_other_pct=tn_2012_pct_county[(tn_2012_pct_county['party']!=" Republican")&(tn_2012_pct_county['party']!=" Democratic")].reset_index(drop=True)
tn_2012_other_pct
```

```python
# sum by county
tn_2012_other_pct=tn_2012_other_pct.groupby(['county'],as_index=False)['pct_county_vote'].sum()
tn_2012_other_pct
```

```python
# rename 'pct_county_vote' to 'other_pct' and drop "party" to prepare for merge 
tn_2012_other_pct=tn_2012_other_pct.rename(columns={'pct_county_vote':'Other'})
tn_2012_other_pct
```

```python
# join tn_2012_rep, tn_2012_dem_pct and tn_2012_other_pct to form tn_2012_county_pres_pct
dem=tn_2012_dem_pct['Democratic']
other=tn_2012_other_pct['Other']
tn_2012_county_pres_pct=tn_2012_rep_pct.join(dem).join(other)
tn_2012_county_pres_pct
```

### Add column for Republican margin over Democrat for heat mapping in Tableau

```python
tn_2012_county_pres_pct['Rep_Margin']= tn_2012_county_pres_pct['Republican']-tn_2012_county_pres_pct['Democratic']
tn_2012_county_pres_pct.head(25)
```
